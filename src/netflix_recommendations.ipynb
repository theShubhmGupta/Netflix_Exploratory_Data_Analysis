## REC 01 — Commit to Multi-Season TV Shows Before Launch

**Notebook Reference:** Plot 03 (TV Show Seasons Countplot) + Outlier Check — TV Seasons

**Finding:**
Over 60% of TV Shows on Netflix have only 1 season. The median season count is 1 and the mean is just 1.76. Only 259 shows out of 2,676 exceed 4 seasons. Long-running shows like Grey's Anatomy (17 seasons), Supernatural (15), and NCIS (15) are clear outliers — not the norm.

**Action:**
Before commissioning or acquiring a new TV Show, commit to a guaranteed 2-3 season run. Audiences invest time in characters and storylines — a cancellation after 1 season creates frustration and directly increases monthly churn.

**Expected Outcome:**
Higher subscriber retention and stronger word-of-mouth for new shows.

## REC 02 — Increase TV Show Investment in South Korea and Japan

**Notebook Reference:** Plot 09 (Countries vs Movies/TV Shows) + Plot 07 (Countries vs Release Year) + Top 10 Countries table (South Korea: 231, Japan: 318)

**Finding:**
South Korea and Japan show a much higher TV Show-to-Movie ratio compared to Western markets. Their content is concentrated in recent release years — meaning these are active, growing markets. Japan (318) and South Korea (231) rank 6th and 8th in total contribution despite being smaller economies than others on the list.

**Action:**
Fund more original TV Show commissions in South Korea (dramas, thrillers) and Japan (anime, slice-of-life series). These genres already have proven global fan bases well beyond their home markets.

**Expected Outcome:**
Subscriber growth across Asia-Pacific and international audiences who actively seek K-dramas and anime.

## REC 03 — Expand India's Contribution from Movies into TV Shows

**Notebook Reference:** Plot 09 (Countries vs Movies/TV Shows) + Top 10 Countries table (India: 1,046 titles — almost entirely Movies)

**Finding:**
India is the second largest content contributor with 1,046 titles, but its output is heavily skewed towards Movies (Bollywood). Indian original TV Shows are significantly underrepresented on the platform compared to what the market can produce and what Indian subscribers want.

**Action:**
Commission more Indian original TV Shows — particularly web series in Hindi and regional languages covering drama, thriller, and comedy. India has a massive and rapidly growing streaming subscriber base that is underserved by the current catalogue.

**Expected Outcome:**
Deeper engagement with Indian subscribers and reduced dependency on Bollywood movie licencing deals.

## REC 04 — Schedule Major Content Drops in January and July

**Notebook Reference:** Plot 11 (Monthly Content Addition — Movies vs TV Shows) 


**Finding:**
January and July are the two peak months for new content additions across both Movies and TV Shows throughout the full dataset history. This pattern is consistent across multiple years. Both months align with natural viewer behaviour spikes — post-holiday re-engagement in January and summer holidays in July.

**Action:**
Plan the launch of flagship original shows and high-budget movies specifically for January and July. Avoid saving major content for the Q3 shoulder months where viewer engagement is naturally lower.

**Expected Outcome:**
Higher launch viewership, stronger social media buzz, and improved subscriber acquisition during natural peak periods.

## REC 05 — Fill the February and August Content Gap

**Notebook Reference:** Plot 11 (Monthly Content Addition)

**Finding:**
February and August consistently have the fewest new titles added across the full dataset. This creates a visible dip in the content calendar. Subscribers who have already consumed the January and July drops will notice the gap — and this is when churn risk is highest.

**Action:**
Shift a portion of mid-tier content releases — acquired films, returning seasons, and international titles — into February and August to smooth out the content calendar and maintain subscriber engagement year-round.

**Expected Outcome:**
Reduced seasonal churn and more consistent monthly active usage across the full year.

## REC 06 — Grow Family and Children's Content to at Least 20% of New Acquisitions

**Notebook Reference:** Plot 10 (Rating Distribution by Type) + Rating Distribution table (TV-Y: 307, TV-G: 220, PG: 287, G: 41 — combined ~10% of catalogue)

**Finding:**
TV-MA accounts for 36% of all titles. Family-friendly ratings (TV-Y, TV-Y7, TV-G, PG, G combined) make up roughly 10% of the entire catalogue. This is a significant gap — household subscriptions include children, and a family that cannot find enough kid-appropriate content has a clear reason to cancel.

**Action:**
Set a deliberate acquisition target of 20% family-rated content in new additions each year. Focus on animated series, live-action family dramas, and children's educational content.

**Expected Outcome:**
Higher retention among household subscription accounts — families with children are among the highest long-term value subscriber segments.

## REC 07 — Prioritise Netflix Originals Over Licenced Third-Party Content

**Notebook Reference:** Plot 15 (Correlation Heatmap) + Plot 16 (Pairplot) + Release Year Distribution (median release year = 2017)

**Finding:**
release_year and year_added show a moderate positive correlation (~0.5). Licenced content typically arrives on Netflix well after its original release — already old news by the time it lands. Back-catalogue content (pre-2010) was added much later. The pairplot confirms this positive trend clearly across both content types.

**Action:**
Shift content budget away from licencing deals towards Netflix original productions that launch directly on the platform on day one. Originals create exclusive, must-watch moments that competitors cannot replicate and audiences cannot find elsewhere.

**Expected Outcome:**
Lower licencing costs, stronger platform differentiation, and content that feels fresh and current at launch.

## REC 08 — Build Dedicated Content Pipelines for Underserved Genres

**Notebook Reference:** Plot 12 (Genre Word Cloud) + Top 10 Genres table

**Finding:**
International Movies, Dramas, and Comedies dominate the genre landscape by a wide margin. Niche genres — anime, stand-up comedy, faith-based content, and investigative documentaries — are visible in the word cloud but noticeably smaller. These genres have dedicated, high-engagement audiences who subscribe specifically for this content.

**Action:**
Commission a consistent quarterly pipeline of niche genre content. Specifically: 4-6 stand-up specials per year, a dedicated anime original slate, and 3-4 investigative documentary series per year. Consistency matters more than volume here.

**Expected Outcome:**
Niche audiences subscribe, watch intensely, and stay retained as long as new content keeps coming. Without a reliable pipeline, they subscribe, finish everything, and cancel.

## REC 09 — Use Mature Content to Acquire, Family Content to Retain

**Notebook Reference:** Plot 05 (Rating vs Movie Duration Boxplot) + Plot 10 (Rating Distribution by Type) + Basic Metrics (TV-MA = 3,207 titles)

**Finding:**
TV-MA is Netflix's single biggest content category at 36% of all titles. Mature-rated movies are also longer on average — meaning they represent bigger production investments and are the primary draw for adult subscribers. Family content makes up only ~10% of the catalogue. Both content types serve different purposes in the subscriber lifecycle.

**Action:**
Continue leading with high-quality adult originals as the primary acquisition driver — these are what make new subscribers sign up. But pair this with a deliberate family content strategy so that once a household subscribes, there is content for every member of the family.

**Expected Outcome:**
A two-tier strategy that maximises both subscriber acquisition (adult originals) and long-term household retention (family content).

## REC 10 — Accelerate Content Acquisition from Emerging Markets

**Notebook Reference:** Plot 07 (Countries vs Release Year) + Plot 08 (Content Growth Over Time) + Top 10 Countries table (France: 393, Spain: 232, Germany: 226, Mexico: 169)

**Finding:**
France, Spain, Germany, and Mexico are growing contributors but still well behind India and the US. Content from these markets is concentrated in recent years — Netflix has only recently started acquiring from them. Plot 08 also confirms that overall content growth has slowed post-2019, making emerging market content a key lever for filling the pipeline.

**Action:**
Establish dedicated acquisition deals with production companies in France, Spain, Mexico, and Brazil. These are large subscriber markets where local-language content is a proven driver of sign-ups and retention. Local audiences are far more likely to subscribe when they see content made for them in their own language.

**Expected Outcome:**
Accelerated subscriber growth in Europe and Latin America — two regions with significant untapped streaming potential.
